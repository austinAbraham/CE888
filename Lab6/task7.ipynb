```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn import cluster, datasets, metrics
import seaborn as sns
```

```
df_country = pd.read_csv('https://raw.githubusercontent.com/sagihaider/CE888_2021/main/Lab_6/Data/Country-data.csv', header=0, index_col=None)
df_country.head(5)
```

```
print("Shape of data:",df_country.shape)
```

```
x_country, x_data = df_country.iloc[:, 0:1], df_country.iloc[:, 1:10]
print(x_country.head(5))
print(x_data.head(5))
```

```
km_country = cluster.KMeans(n_clusters=3)
km_country.fit(x_data)
print(km_country.labels_)
```

```
df_labelled = df_country.copy()
df_labelled['label'] = km_country.labels_
print(df_labelled.head(5))
df_labelled['label'].value_counts()
```

```
df_labelled[['label']] = df_labelled['label'].replace([0,1,2], ['under-developing country', 'developing country', 'developed country'])
df_labelled[df_labelled['label'] == 'developed country']
```
